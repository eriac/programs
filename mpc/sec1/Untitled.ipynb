```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
step=100
dt=0.1
t=np.arange(0,step*dt,dt)

m=1.0

x=np.zeros((step,2))
u=np.zeros(step)
x[0,0]=1

A=np.array([
    [0,1],
    [0,0]
])
B=np.array([
    [0,1/m]
])

for i in range(step-1):
    # control
    kp=0.1
    ki=0
    kd=0.1
    u[i]=-kp*x[i,0]-ki*np.sum(x[0:i+1,0])-kd*x[i,1]
    
    # model
    x[i+1,:]=x[i,:]+np.dot(A,x[i,:])*dt+B*u[i]

print(t.shape, x[:,0].shape)
plt.plot(t, x[:,0],"r")
plt.plot(t, x[:,1],"g")
plt.plot(t, u[:],"b")

plt.show()
```
